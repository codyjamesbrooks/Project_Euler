```python
# Problem #11
# What is the greatest product of four adjacent numbers in the same direction
# (up, down, left, right, or diagonally) in the 20 X 20 grid provided. 

data = [
'08 02 22 97 38 15 00 40 00 75 04 05 07 78 52 12 50 77 91 08',
'49 49 99 40 17 81 18 57 60 87 17 40 98 43 69 48 04 56 62 00',
'81 49 31 73 55 79 14 29 93 71 40 67 53 88 30 03 49 13 36 65',
'52 70 95 23 04 60 11 42 69 24 68 56 01 32 56 71 37 02 36 91',
'22 31 16 71 51 67 63 89 41 92 36 54 22 40 40 28 66 33 13 80',
'24 47 32 60 99 03 45 02 44 75 33 53 78 36 84 20 35 17 12 50',
'32 98 81 28 64 23 67 10 26 38 40 67 59 54 70 66 18 38 64 70',
'67 26 20 68 02 62 12 20 95 63 94 39 63 08 40 91 66 49 94 21',
'24 55 58 05 66 73 99 26 97 17 78 78 96 83 14 88 34 89 63 72',
'21 36 23 09 75 00 76 44 20 45 35 14 00 61 33 97 34 31 33 95',
'78 17 53 28 22 75 31 67 15 94 03 80 04 62 16 14 09 53 56 92',
'16 39 05 42 96 35 31 47 55 58 88 24 00 17 54 24 36 29 85 57',
'86 56 00 48 35 71 89 07 05 44 44 37 44 60 21 58 51 54 17 58',
'19 80 81 68 05 94 47 69 28 73 92 13 86 52 17 77 04 89 55 40',
'04 52 08 83 97 35 99 16 07 97 57 32 16 26 26 79 33 27 98 66',
'88 36 68 87 57 62 20 72 03 46 33 67 46 55 12 32 63 93 53 69',
'04 42 16 73 38 25 39 11 24 94 72 18 08 46 29 32 40 62 76 36',
'20 69 36 41 72 30 23 88 34 62 99 69 82 67 59 85 74 04 36 16',
'20 73 35 29 78 31 90 01 74 31 49 71 48 86 81 16 23 57 05 54',
'01 70 54 71 83 51 54 69 16 92 33 48 61 43 52 01 89 19 67 48'
]

# store the data in a list on a [y][x] ordered pair system. With the orign
# (0,0) being the top left number
for counter in range(len(data)):
    data[counter] = data[counter].split(' ')
for i in range(0,20):
    for j in range(0,20):
        data[i][j] = int(data[i][j])
    
greatest_product = 0
# The method I chose to pursue is checking each direction indivdualy, 
# always updating 'greatest_product' 
for row in data:
    for i in range(0,17):
        product = 1
        for num in row[i:i+4]:
            product *= num
            if product > greatest_product:
                greatest_product = product
            
for i in range(0,17):
    for j in range(0,20):
        product = data[i][j] * data[i+1][j] * data[i+2][j] * data[i+3][j]
        if product > greatest_product:
            greatest_product = product

for i in range(0,17):
    for j in range(0,17):
        product = data[i][j] * data[i+1][j+1] * data[i+2][j+2] * data[i+3][j+3]
        if product > greatest_product:
            greatest_product = product
    
for i in range(0, 17):
    for j in range(19, 2, -1):
        product = data[i][j] * data[i+1][j-1] * data[i+2][j-2] * data[i+3][j-3]
        if product > greatest_product:
            greatest_product = product
    

print(f'The solution is {greatest_product}.')        
```

```python
# Problem 12
# Highly divisible triangular number
# The sequence of triangle numbers is generated by adding the natural numbers. 
# So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. The first ten terms would be:

# 1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...

# Let us list the factors of the first seven triangle numbers:

#  1: 1
#  3: 1,3
#  6: 1,2,3,6
# 10: 1,2,5,10
# 15: 1,3,5,15
# 21: 1,3,7,21
# 28: 1,2,4,7,14,28
# We can see that 28 is the first triangle number to have over five divisors.

# What is the value of the first triangle number to have over five hundred divisors?
from functools import reduce

def factors(n):
    return len(set(reduce(list.__add__,([i, n//i] for i in range(1, int(n**.5)+1) if n % i == 0))))

triangular_numbers = []
n = 1
div = 0

while div <= 500:
    tri = sum(range(1, n+1))
    triangular_numbers.append(tri)
    div = factors(tri)
    n += 1
print(f'The solution is {triangular_numbers[-1]}.')
```

```python
# Problem 13
# Work out the first ten digits of the sum of the following one-hundred 50
# digit numbers. 

# I stored the numbers in the list data.For the sake of space I have not 
# printed data. 
# Solution method involved summing progressivly larger slices of the first n
# digits of the numbers. And comparing the change in that sum over time. 
# Once our solution achieved some amount of stability.
answer = ''
for n in range(10, 20):
    approx_sum = 0
    for number in data:
        temp = str(number)[0:n+1]
        approx_sum += int(temp)
    if str(approx_sum)[:10] != answer:
        answer = str(approx_sum)[0:10]
print('The solution is {}'.format(answer))    
```

```python
# Problem 14

# The following iterative sequence is defined for the set of positive integers:

# n → n/2 (n is even)
# n → 3n + 1 (n is odd)

# Using the rule above and starting with 13, we generate the following
# sequence:

# 13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
# It can be seen that this sequence (starting at 13 and finishing at 1) 
# contains 10 terms. Although it has not been proved yet (Collatz Problem),
# it is thought that all starting numbers finish at 1.

# Which starting number, under one million, produces the longest chain?
def seq(n, i=1):
    while n > 1:
        if n % 2 == 0:
            i += 1
            n = int(n/2)
        else:
            i += 1
            n = 3 * n + 1
    return i

largest_count = 0
start_num = 0

for x in range(1, 1000001):
    count = seq(x)
    if count > largest_count:
        largest_count = count
        start_num = x
print(f'The solution to the problem is {start_num}.')
```

```python
# Problem 17
# If the numbers 1 to 5 are written out in words: one, two, three, four,
# five, then there are 3 + 3 + 5 + 4 + 4 = 19 letters used in total.

# If all the numbers from 1 to 1000 (one thousand) inclusive were written
# out in words, how many letters would be used?
num_dict = {
    1: 'one',
    2: 'two',
    3: 'three', 
    4: 'four', 
    5: 'five', 
    6: 'six', 
    7: 'seven',
    8: 'eight',
    9: 'nine',
    10: 'ten', 
    11: 'eleven', 
    12: 'twelve', 
    13: 'thirteen',
    15: 'fifteen',
    18: 'eighteen',
    20: 'twenty', 
    30: 'thirty',
    40: 'forty', 
    50: 'fifty', 
    60: 'sixty', 
    70: 'seventy', 
    80: 'eighty',
    90: 'ninety'}
alpha = 'abcdefghijklmnopqrstuvwxyz'
def spell_it_out(n):
    # given int n, returns n completly spelled out.
    # Well almost completly. I didn't include any spaces
    # or hyphens. She everything is one string.  
    if str(n) == '00' or str(n) == '0':
        return ""
    elif len(str(n)) == 1:
        return num_dict[n]
    elif len(str(n)) == 2:
        if n in num_dict.keys():
            return num_dict[n]
        else:
            first = int(str(n)[0])
            second = int(str(n)[1])
            if first == 1:
                return num_dict[second] + 'teen'
            else:
                first = int(str(first) + '0')
                return num_dict[first] + num_dict[second]
    elif len(str(n)) == 3:
        first = int(str(n)[0])
        rest = int(str(n)[1:])
        if rest == 0:
            return num_dict[first] + 'hundred'
        else:
            return num_dict[first] + 'hundredand' + spell_it_out(rest)
    else:
        return 'onethousand'

count = 0
for i in range(1, 1001):
    count += len(spell_it_out(i))
print(f'The solution to the problem is {count}.')
```

```python
# problem 18
# Find the maximum total from top to bottom of the triangle below:
# First I stored the triangel in a list. Where each item was a different row 
# in the triangle. 
data = [
'75',
'95 64',
'17 47 82',
'18 35 87 10',
'20 04 82 47 65',
'19 01 23 75 03 34',
'88 02 77 73 07 63 67',
'99 65 04 28 06 16 70 92',
'41 41 26 56 83 40 80 70 33',
'41 48 72 33 47 32 37 16 94 29',
'53 71 44 65 25 43 91 52 97 51 14',
'70 11 33 28 77 73 17 78 39 68 17 57',
'91 71 52 38 17 14 91 43 58 50 27 29 48',
'63 66 04 68 89 53 67 30 73 16 69 87 40 31',
'04 62 98 27 23 09 70 98 73 93 38 53 60 04 23']

for counter in range(len(data)):
    data[counter] = data[counter].split(' ')
# Then I decided on a brute force approach, to check every possible path down
# the triangle. To do this I generated a list of all possible combinations of
# 14 different 0's (choose left num) or 1's (choose right num)
from itertools import product
combos = list(product(range(2), repeat=14))

# Then I found the sum for each path, and updated answer, with the largest
# sum each time
answer = 0
for path in combos:
    path_total = 75 # Top number, part of every path
    cur_pos = 0 # Restart at top of triange, will be used for lateral mvmt
    cur_tier = 1 # Restat at second layer, used for up down mvmt. 
    for char in path:
        if char == 0:
            path_total += int(data[cur_tier][cur_pos]) # Update sum with left num
        else: 
            path_total += int(data[cur_tier][cur_pos+1])
            cur_pos += 1 
        cur_tier += 1 # work down the triangle
    answer = max(answer, path_total)
print(f'The solution to the problem is {answer}.')
```

```python
# Problem 19
# You are given the following information, but you may prefer to do some research
# for yourself.

# 1 Jan 1900 was a Monday.

# Thirty days has September,
# April, June and November.
# All the rest have thirty-one,
# Saving February alone,
# Which has twenty-eight, rain or shine.
# And on leap years, twenty-nine.

# A leap year occurs on any year evenly divisible by 4, but not on a century
# unless it is divisible by 400. 

# How many Sundays fell on the first of the month during the twentieth century
# (1 Jan 1901 to 31 Dec 2000)?
import datetime

sunday_count = 0

for x in range(1901, 2001): # Check every year between 1901 and 2000
    for y in range(1, 13): # Check every month of every year
        day = datetime.datetime(x, y, 1) # get a datetime for every day in the
                                         # 20th century
        if day.strftime('%A') == 'Sunday':
            sunday_count += 1
print(f'The solution to the problem is {sunday_count}.')
```

```python
# Problem 20
# n! means n × (n − 1) × ... × 3 × 2 × 1

# For example, 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800,
# and the sum of the digits in the number 10! is 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27.

# Find the sum of the digits in the number 100!
# Using standard lib
product = 1
for x in range(100, 0, -1):
    product *= x
answer = sum(int(x) for x in str(product))
print(f'The solution is {answer}')

# Did a reduce solution just for the heck of it. 
from functools import reduce
def mul(x,y): 
    return x*y
def add(x,y):
    return int(x)+int(y)
def choose(n):
    return reduce(mul, range(n, 0, -1))
print(f'The solution is {reduce(add, str(choose(100)))}.')
```
